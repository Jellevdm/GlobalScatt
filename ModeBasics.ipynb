```python
import numpy as np
import os
from matplotlib import pyplot as plt
from drama import utils as drtls
from drama.geo import QuickRadarGeometry, timing_diagram, mode_on_timing
from drama import constants as const

#ADDED
from drama.io import cfg
```

```python
#Import configuration file - Added
main_dir = 'C:/Users/Jelle/Desktop/Msc Space Engineering/Earth Observation/Radar Assignment'
pltdir = 'C:/Users/Jelle/Desktop/Msc Space Engineering/Earth Observation/Radar Assignment/plots'
par_file = os.path.join(main_dir, "GlobalScat.cfg")
conf = cfg.ConfigFile(par_file)
```

# Timing diagram

```python
# Orbit A1
#Extract orbit/payload parameters - Added
orbit_name = "orbit_A1"             #Change to obtain different orbit from config
payload_name = "global_scat"
orbit = getattr(conf, orbit_name)
payload = getattr(conf, payload_name)

h_orb = orbit.Horb
k = orbit.days_cycle
j = orbit.orbits_nbr
angle_range = orbit.angle_range
La = payload.La
swath_min = orbit.swt_min
print(f"Minimum swath - global coverage:{swath_min} meter")

f0 = 5.4e9 # C-band
wl = 3e8 / f0
print(f"Wavelength: {wl:.2f} m")
eta_a = 0.7
#beamwidth
theta_a = wl / (eta_a * La)
print(f"Beamwidth: {theta_a:.2f} rad")
qgeo = QuickRadarGeometry(h_orb)

incsd = np.linspace(angle_range[0], angle_range[1])
incsr = np.radians(incsd)
td = timing_diagram([incsd.min(), incsd.max()], h_orb, np.linspace(100,2500), 0.1)

# For determining the available areas in the plot
mPRFs = [1000] #, 8250, 8800, 9125, 8250, 8820]
mth0= [25] #, 26.7,29.8, 32.1, 33.5, 35.2]
mswth = [40e3] #, 25e3, 24e3, 22e3, 21e3, 20e3]
for prf, th0, swth in zip(mPRFs, mth0, mswth):
    mode_on_timing(prf, th0, swth, td)
plt.savefig(os.path.join(pltdir,'timing.png'), dpi=300)
```
